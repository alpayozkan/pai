```python
import os
import typing
from sklearn.gaussian_process.kernels import *
import numpy as np
from sklearn.gaussian_process import GaussianProcessRegressor
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
train_x = np.loadtxt('train_x.csv', delimiter=',', skiprows=1)
train_y = np.loadtxt('train_y.csv', delimiter=',', skiprows=1)
# test_x = np.loadtxt('test_x.csv', delimiter=',', skiprows=1)
```

```python
p = 0.7
mask = np.random.binomial(1, p, train_x.shape[0]).astype(np.bool)
```

```python
(mask==1).sum()
```

```python
(mask==0).sum()
```

```python
x_train = train_x[mask]
x_test = train_x[np.logical_not(mask)]

y_train = train_y[mask]
y_test = train_y[np.logical_not(mask)]
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
# swap for the rest of the code,

train_x = x_train
test_x = x_test

train_y = y_train
test_y = y_test
```

```python
from sklearn.cluster import KMeans
```

```python
num_clust = 15
alpha = 0.1


scaled_train_x = train_x.copy()
scaled_train_x[:,2] = alpha*scaled_train_x[:,2]

kmeans = KMeans(n_clusters=num_clust, random_state=0, n_init="auto").fit(scaled_train_x)
```

```python
kmeans.labels_
```

```python
kmeans.predict(scaled_train_x)
```

```python
np.sum((kmeans.predict(scaled_train_x)!=kmeans.labels_))
```

```python
kmeans.predict(train_x)
```

```python
cmeans = kmeans.cluster_centers_
```

```python
plt.figure(figsize=(20, 20), dpi=80)
plt.scatter(np.array(train_x[:,0]), np.array(train_x[:,1]), c = np.array(train_x[:,2])) 
plt.colorbar(label="Like/Dislike Ratio", orientation="horizontal") 

plt.scatter(cmeans[:,0], cmeans[:,1], c = cmeans[:,2], cmap='Reds', marker='x')
plt.colorbar(label="Like/Dislike Ratio", orientation="horizontal") 

```

```python
kernel = Matern(length_scale_bounds=(1e-3, 1e5))
gprlist = [GaussianProcessRegressor(kernel=kernel, random_state=0) for i in range(num_clust)]
```

```python
for idx in range(num_clust):
    gpr = gprlist[idx]
    
    indx_list = np.where(kmeans.labels_==idx)[0]
    x_nb = train_x[indx_list]
    y_nb = train_y[indx_list]
    
    gpr.fit(x_nb, y_nb)
```

```python
scores = []
samples = []
```

```python
for idx in range(num_clust):
    gpr = gprlist[idx]
    
    indx_list = np.where(kmeans.labels_==idx)[0]
    x_nb = train_x[indx_list]
    y_nb = train_y[indx_list]
    
    score = gpr.score(x_nb, y_nb)
    scores.append(score)
    samples.append(x_nb.shape[0])
```
